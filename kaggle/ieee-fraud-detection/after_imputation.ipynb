```python
import pandas as pd
import numpy as np
imputedData = pd.read_csv("data/imputed_data.csv", index_col = "TransactionID")
imputedData.head()
```

```python
y = imputedData["isFraud"].values
print("First 10 of y = ", y [0:10])

X = imputedData.drop("isFraud", axis = 1).values
print("\nFirst 10 by 10 of X = \n", X [0:10, 0:10])
```

```python
# learning
# from sklearn.neural_network import MLPClassifier
# from sklearn.preprocessing import OneHotEncoder
# classifier = MLPClassifier()
# encoder = OneHotEncoder
# X = encoder.fit_transform(X)
# classifier.fit (X, y)
# perhaps, it failed because I fortot brackets, when constructing the encoder
```

```python
# learning
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 100)
# classifier.fit (X, y)
# theoretically it can work with categorical data. But in reality it is implemented to work with numbers
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
Xtransformed = encoder.fit_transform(X)

```

```python
# predictiing
# divide into training and testing sets
from sklearn.model_selection import train_test_split
Xtrain, Xtest, yTrain, yTest = train_test_split(Xtransformed, y, test_size = 0.3)
classifier.fit (Xtrain, yTrain)
```

```python
# testing
from sklearn.metrics import accuracy_score
yPredicted = classifier.predict(Xtest)

# check by random choice
import numpy as np
arrayLength = yPredicted.shape[0]
print("arrayLength =", arrayLength)
yFirst5ones = np.zeros(arrayLength)
yFirst5ones [0:5] = 1
print (accuracy_score(y_true = yTest, y_pred = yPredicted))
print ("sum(yPredicted) =", sum(yPredicted))
print ("sum(yTest) =", sum(yTest))

print (accuracy_score(y_true = yTest, y_pred = yFirst5ones))
print ("sum(yFirst5ones) =", sum(yFirst5ones))
print ("sum(yTest) =", sum(yTest))

```
